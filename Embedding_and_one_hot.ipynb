```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
sent
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=10000    # Vocabulary size parameter is important for one_hot
```

```python
one_hot_rep=[one_hot(word,voc_size) for word in sent]
one_hot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential   # Sequential is required for embedding layer.
from tensorflow.keras.preprocessing.sequence import pad_sequences   # Pad sequences is required for passing input.
```

```python
model=Sequential()
```

```python
model.add(Embedding(voc_size,10,input_length=8))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(pad_sequences(one_hot_rep,maxlen=8,padding='pre'))
```
